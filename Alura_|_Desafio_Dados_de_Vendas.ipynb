```python
import pandas as pd
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df = pd.DataFrame(vendas_2022, index=lojas)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2,2, figsize=(13,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas de 2022 das lojas A, B, C e D')

axs[0,0].plot(df.loc['A', vendas_2022])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df.loc['B', vendas_2022])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df.loc['C', vendas_2022])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df.loc['D', vendas_2022])
axs[1,1].set_title('Loja D')


for ax in axs.flat:
  ax.set_xlabel('Mês')
  ax.set_ylabel('Número de Vendas')

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```
